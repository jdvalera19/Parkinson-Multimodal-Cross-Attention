```python
import torchaudio

import numpy             as np
import matplotlib.pyplot as plt

from scipy import signal
```

```python
audio = '/home/brayan/AudioVisualData_v2/Control/C4/Vowels/audio/C4-1-A.mp3'
```

```python
sig, sr = torchaudio.load(audio)

sig_mid = len(sig[0])//2

sig = sig[0, sig_mid-2400:sig_mid+2400]


f, t, Sxx = signal.spectrogram(sig,fs=48000.,nperseg=480,noverlap=240,nfft=512)

Sxx = 10 * np.log10(Sxx + 1e-7)
```

```python
plt.pcolormesh(t, f, Sxx, cmap = 'inferno')
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```

```python
plt.specgram(sig, Fs=sr, cmap='inferno')
plt.axis('off')
```
